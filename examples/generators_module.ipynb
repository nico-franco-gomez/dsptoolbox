# `generators`

The generators module contains some functions that create signals useful to analyze different systems. These are:

- noise (white, pink, etc.)
- harmonic
- dirac
- chirp

```python
import dsptoolbox as dsp
```

### Create signal

```python
n = dsp.generators.noise(type_of_noise='white', sampling_rate_hz=5_000)
n.plot_magnitude()
n = dsp.generators.noise(type_of_noise='blue', sampling_rate_hz=5_000)
n.plot_magnitude();
```

### Multi-channel sine tone

```python
sin = dsp.generators.harmonic(frequency_hz=10, sampling_rate_hz=300,
                              number_of_channels=3, uncorrelated=True)
sin.plot_time();
```
